```python
from sir import SIR
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
options = {
    "N_max": 1000001,
    "N_step": 1000000,
    "N_min": 1000000,
    "beta_max": 0.3,
    "beta_min": 0.2,
    "beta_step": 0.1,
    "gama_max": 11,
    "gama_min": 10,
    "gama_step": 10,
    "t_max": 301,
    "t_min": 300,
    "t_step": 100,    
}
```

```python
sir = SIR()
```

```python
(X, Y) = sir.get_train_data(options)
sir.compile(optimizer=tf.keras.optimizers.Adam())
callback = tf.keras.callbacks.ModelCheckpoint(filepath="models/sir_ck.h5",
                                              monitor='mae',
                                              save_best_only=True)

history = sir.fit(X=X,
                  Y=Y,
                  batch_size=10,
                  epochs=1000,
                  verbose=0,
                  callbacks=[callback])
sir.save()
sir.plot_history(history)
```

```python
# Theoretical modell
N = 1000000
I0, R0 = 1, 0
S0 = N - I0 - R0
y0 = S0, I0, R0
beta, gamma = 0.2, 1./10
t = np.linspace(0, 300, 300)
data = sir.get_data_from_intervall(y0=y0,
                                   t_ls=t,
                                   args=(N, beta, gamma)) 
data = tf.keras.utils.normalize(data)
S, I, R = data.T
```

```python
sir.from_pretrained()
sir.pretrained.summary()
X = sir.fill_X(t, beta, gamma)
X = tf.keras.utils.normalize(X)
pred = sir.pretrained.predict(X)
SA, IA, RA = pred.T
```

```python
fig = plt.figure(facecolor='w', figsize=(15,9))
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S, 'b', alpha=0.5, lw=1, label='Susceptible T model')
ax.plot(t, SA, 'b', alpha=1, lw=1, label='Susceptible A model')
ax.plot(t, I, 'r', alpha=0.5, lw=1, label='Infected T model')
ax.plot(t, IA, 'r', alpha=1, lw=1, label='Infected A model')
ax.plot(t, R, 'g', alpha=0.5, lw=1, label='Recovered T model')
ax.plot(t, RA, 'g', alpha=1, lw=1, label='Recovered A model')

ax.set_xlabel('Time /days')
ax.set_ylabel('Number')
#ax.set_ylim(0,5)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```
